```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("sales_team.csv")
```

```python
df=df.drop(['User ID','Gender'],axis=1)
```

```python
k=int(input("enter value for folds(k) : "))
```

```python
X=df.drop('Purchased',axis=1)
y=df['Purchased']
```

```python
X.head()
```

```python
n=len(df)
print(f'n = {n}')
start=0
end=int(n/k)
start_list=[]
end_list=[]
for i in range(k):
    start_list.append(start)
    end_list.append(end)
    print(f"start = {start} and end = {end}")
    start=int(((i+1))*(n/k))
    end=int(((i+2)*(n/k)))
start_list,end_list
```

### Below cell is just for understanding purposes

```python
for i in range(k-1):
    print('TEST  => start_list[i] = ',start_list[i],' end_list[i] = ',end_list[i])
    print('TRAIN => start_list[i+1] = ',start_list[i+1],' end_list[i+1] = ',end_list[k-1])
    if(i!=0):
        print('TRAIN + => ',start_list[0]," : ",start_list[i])
    print('**'*10)
    
#     test=X[start_list[i]:end_list[i]]
#     train=X[start_list[i+1]:end_list[i+1]]
```

```python
lr=LogisticRegression(max_iter=1000)
acc_list=[]
```

```python
for i in range(k-1):
    test=None
    train=None
    test=df[start_list[i]:end_list[i]]
    train=df[start_list[i+1]:end_list[k-1]]
    print('TEST  =>  = ',start_list[i], '  : ',end_list[i])
    print('TRAIN =>  = ',start_list[i+1],' :  ',end_list[k-1])
    if(i!=0):
        print('TRAIN + => ',start_list[0]," : ",start_list[i])
        train = train.append(df[start_list[0]:start_list[i]])
    X_train,X_test=train.drop('Purchased',axis=1),test.drop('Purchased',axis=1)
    y_train,y_test=train['Purchased'],test['Purchased']
    print(f"len of x = {len(X_train)}, len of y = {len(y_train)}")
    sc=StandardScaler()
    X_train,X_test=sc.fit_transform(X_train),sc.fit_transform(X_test)
    lr.fit(X_train,y_train)
    y_pred = lr.predict(X_test)
    print("accuracy of model = ",accuracy_score(y_test,y_pred))
    acc_list.append(accuracy_score(y_test,y_pred))
    print('**'*10)

    
```

```python
print('Accuracy list = ',acc_list)
```

```python
print(f'Average accuracy  = {sum(acc_list)/len(acc_list)}')
```
